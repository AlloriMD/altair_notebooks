Google's stock price over time.

## Load Dataset
The data comes in the form of a Pandas Dataframe:

```python
from altair import load_dataset
data = load_dataset("stocks")
data.head()
```

## Define Altair Specification

```python
from altair import *  # Import the altair API

chart = Chart(data).mark_point().encode(
    color=Color('date:T',
        legend=Legend(
            format='%y',
            values=[DateTime(
                    year=2004.0,
                ), DateTime(
                    year=2007.0,
                ), DateTime(
                    year=2010.0,
                )],
        ),
        scale=Scale(
            domain=[DateTime(
                    year=2004.0,
                ), DateTime(
                    year=2010.0,
                )],
        ),
    ),
    x=X('date:T',
        axis=Axis(
            values=[DateTime(
                    year=2005.0,
                ), DateTime(
                    year=2007.0,
                ), DateTime(
                    year=2009.0,
                )],
        ),
        scale=Scale(
            domain=[DateTime(
                    month=1.0,
                    year=2005.0,
                ), DateTime(
                    year=2009.0,
                )],
        ),
    ),
    y='price:Q',
).transform_data(
    filter="datum.symbol==='GOOG'",
)

chart.max_rows = len(data)  # Altair 1.2.1 has a max_rows of 500 by default
```
