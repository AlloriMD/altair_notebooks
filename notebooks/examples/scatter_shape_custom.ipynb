## Load Dataset
The data comes in the form of a Pandas Dataframe:

```python
from altair import load_dataset
data = load_dataset("cars")
data.head()
```

## Define Altair Specification

```python
from altair import *  # Import the altair API

chart = Chart(data).mark_point(
    shape='M0,0.2L0.2351,0.3236 0.1902,0.0618 0.3804,-0.1236 0.1175,-0.1618 0,-0.4 -0.1175,-0.1618 -0.3804,-0.1236 -0.1902,0.0618 -0.2351,0.3236 0,0.2Z',
).encode(
    color='Cylinders:N',
    size='Weight_in_lbs:Q',
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
)

chart.max_rows = len(data)  # Altair 1.2.1 has a max_rows of 500 by default
```
