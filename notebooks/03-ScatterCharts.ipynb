# Scatter Chart Examples

```python
import altair as alt
```

```python
cars = alt.load_dataset('cars')
```

## Dot chart

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
)
```

## Scatter chart

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon'
)
```

## Bubble chart

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    size='Acceleration'
)
```

## Scatter chart with color and/or shape

```python
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    shape='Origin',
)
```

## Circle scatterchart

```python
alt.Chart(cars).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```

## Square scatter chart

```python
alt.Chart(cars).mark_square().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```

## Text scatter chart

```python
alt.Chart(cars).mark_text().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    text='Origin'
)
```

## Tick chart

```python
alt.Chart(cars).mark_tick().encode(
    x='Horsepower:Q',
    y='Cylinders:O',
)
```
