# Heatmaps

A heatmap is a rectangular grid with the rectangles within the grid colored to visually encode a numerical value. Making a heatmap in Vega-Lite 1.0 is a bit subtle, so we show examples here.

## Imports

```python
import altair as alt
import pandas as pd
```

```python
# Uncomment/run this line to enable Altair in JupyterLab/nteract:
# alt.enable_mime_rendering()
```

## Load dataset

```python
cars = alt.load_dataset('cars')
```

## Default heatmap

```python
alt.Chart(cars).mark_text().encode(
    alt.Color('mean(Miles_per_Gallon):Q', sort='descending'),
    column='Cylinders:O',
    text='mean(Miles_per_Gallon):Q',
    row='Origin'
)
```

## Coloring background

```python
alt.Chart(cars).mark_text(applyColorToBackground=True).encode(
    alt.Row('Origin:O'),
    alt.Column('Cylinders:O'),
    alt.Color('mean(Miles_per_Gallon):Q', sort='descending'),
    alt.Text('mean(Miles_per_Gallon):Q')
)
```

## Customizing colormap
Vega-lite currently only supports colormaps between two colors.

HTML color names can also be used (e.g., `'#d62728'`).

```python
alt.Chart(cars).mark_text(applyColorToBackground=True).encode(
    alt.Row('Origin:O'),
    alt.Column('Cylinders:O'),
    alt.Color('mean(Miles_per_Gallon):Q', sort='descending', scale=alt.Scale(range=['pink', 'red'])),
    alt.Text(' '),
)
```
