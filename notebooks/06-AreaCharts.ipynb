# Area Charts

```python
import altair as alt
```

```python
# Uncomment/run this line to enable Altair in JupyterLab/nteract:
# alt.enable_mime_rendering()
```

```python
data = alt.load_dataset('unemployment-across-industries')
data['date'] = data['date'].astype(str)
```

## Simple area chart

```python
alt.Chart(data).mark_area().encode(
    alt.X('date:T', axis=alt.Axis(format='%Y', labelAngle=0)),
    y='sum(count)'
)
```

## Stacked area chart

```python
alt.Chart(data).mark_area().encode(
    alt.X('date:T',
      timeUnit='yearmonth',
      axis=alt.Axis(format='%Y', labelAngle=0, title='year')),
    alt.Y('sum(count)'),
    alt.Color('series:N', scale=alt.Scale(range="category20b"))
)
```

## Normalized area chart

```python
alt.Chart(data).mark_area(stacked='normalize').encode(
    alt.X('date:T', timeUnit='yearmonth',
      axis=alt.Axis(format='%Y', labelAngle=0, title='year')),
    alt.Y('sum(count)'),
    color=alt.Color('series:N', scale=alt.Scale(range="category20b"))
)
```

## Streamgraph

```python
alt.Chart(data).mark_area(stacked='center').encode(
    alt.X('date:T', timeUnit='yearmonth',
      axis=alt.Axis(format='%Y', labelAngle=0, title='year')),
    alt.Y('sum(count)'),
    color=alt.Color('series:N', scale=alt.Scale(range="category20b"))
)
```
