```python
data = alt.load_dataset('stocks')
```

## Simple line chart

### Data transform in Vega-Lite

```python
alt.Chart(data).mark_line().encode(
    x='date:T',
    y='price:Q'
).transform_data(filter="datum.symbol==='GOOG'")
```

### Data transform in Pandas

```python
alt.Chart(data.query('symbol == "GOOG"')).mark_line().encode(
    x='date:T',
    y='price:Q'
)
```

### Log scale

```python
alt.Chart(data.query('symbol == "GOOG"')).mark_line().encode(
    y=alt.Y('price:Q', scale=alt.Scale(type='log')),
    x='date:T',
)
```

## Grouped line chart

```python
alt.Chart(data).mark_line().encode(
    x='date:T',
    y='price',
    color='symbol',
)
```

## Line chart with custom path

```python
data = alt.load_dataset('driving')
alt.Chart(data).mark_line().encode(
    alt.X('miles', scale=alt.Scale(zero=False)),
    alt.Y('gas', scale=alt.Scale(zero=False)),
    path='year:T',
)
```
